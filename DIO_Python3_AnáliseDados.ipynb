#Python para análise de dados (Pandas)#

```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
# df = dataframe, pra armazenar o conjunto de dados
# pra ler o conjunto de dados csv
df = pd.read_csv('/content/drive/MyDrive/DIO-Python-AnáliseDados/Datasets/Gapminder.csv', error_bad_lines=False, sep=";")
```

```python
# Visualizando as 5 primeiras linhas
df.head()
```

```python
# Pra trocar os nomes das colunas
df = df.rename(columns={'country':'Pais', 'continent': 'Continente', 'year':'Ano', 'lifeExp':'Expectativa de vida', 'pop':'População total', 'gdpPercap':'PIB'})
```

```python
df.head()
```

```python
# pra retornar as 15 ultimas linhas
df.tail(15)
```

```python
# retorna informações estatisticas do conjunto de dados
df.describe()
```

```python
# retornar apenas valores unicos (unique) da coluna continente
df['Continente'].unique()
```

```python
# retornar apenas os valores do continente Oceania
# criou uma variavel para atribuir a ela
Oceania = df.loc[df['Continente'] == 'Oceania']
Oceania.head()
```

```python
# todo mundo que ta dentro da variavel Oceania, pertence ao continente oceania
Oceania['Continente'].unique()
```

```python
# Pra agrupar dados (group by)
# qtos paises tem por cada continente
# nunique: contagem distinta
df.groupby('Continente')['Pais'].nunique()
```

```python
# qual a expectativa de vida media para cada ano, por ano qual a expectativa de vida media
df.groupby('Ano')['Expectativa de vida'].mean()
```

```python
# media do PIB
df['PIB'].mean()
```

```python
# soma dos PIB's
df['PIB'].sum()
```
